```python
ev_stas = ev_stats.set_index("State")
```

```python
columns = ev_stats.columns[2:-1]
columns
```

###EDA

```python
count = 2
for z in columns:
  sns.barplot(x="State",y=z,data=ev_stats,order=ev_stats.sort_values(z).State)
  sns.color_palette("pastel")
  plt.xticks(rotation=90)
  plt.ylabel('')
  plt.title(f"{z}, hightest = {plt.xticks()[1][-1]}")
  plt.savefig(f"output-{count}.jpg",bbox_inches='tight')
  count+=1
  plt.show()
```

```python
sns.barplot(x="State",y='Total in state',data=ev_stats,order=ev_stats.sort_values("Total in state").State)
sns.color_palette("husl")
plt.xticks(rotation=90)
plt.title(f"Total ev vehicles, hightest = {plt.xticks()[1][-1]}")
plt.savefig("total.jpg",bbox_inches='tight')
plt.show()
```

###Conclusion:
So, from the above results, we can infer that top 4 states that use electric vehicles are Maharashtra, Uttar Pradesh, Gujarat and Delhi.\
This also means that the competition around these states is high enough, and also people of that particular states are showing interest towards the electronic vehicles.

###EDA

```python
plt.figure(1,figsize=(15,7))
n = 0
for cols in ['Age','No of Dependents','Salary','Wife Salary','Total Salary','Price']:
  n += 1
  plt.subplot(1,6,n)
  sns.set(style = 'whitegrid')
  plt.subplots_adjust(hspace=0.5,wspace=0.5)
  sns.violinplot(x= cols, y = 'Make', data=customers_behaviour)
  plt.ylabel("Make" if n==1 else '')
  plt.title('Violin Plot')
plt.savefig("TEst.jpg")
```

###Conclusion:
The people with lower age tend to purchase low cost cars, as the salary increases they tend to purchase costlier cars.\
We could also determine that the increase in dependents would like to purchase a car containing huge space, as a result SUV were purchased more with more dependents.\
If the wife also works, and has a good pay-check, people tend to purchase more costlier cars as the income flow is more compared to the housewives.
